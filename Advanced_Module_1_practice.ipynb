```python
import pandas as pd
df = pd.read_csv('companies.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='state', data=df, palette='viridis')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Count of Companies by State')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x= 'job_openings', y= 'Industry', data=df)
plt.xlabel('Job openings')
plt.ylabel('Industry')
plt.title('Sum of Total Job Openings in Each Industry')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['founded'], df['growth_percentage'])
plt.show()
```

```python
df.sort_values(by='founded', inplace= True)

plt.figure(figsize=(10, 6))
plt.plot(df['founded'], df['growth_percentage'])
plt.show()
```

```python
# Creates new variable/dataframe which shows the average growth_percentage for each founded year
founded_df = df.groupby('founded', as_index=False).agg({'growth_percentage': 'mean'})
founded_df
```

```python
# Shows the average growth rate for each founded year
plt.figure(figsize=(10, 6))
plt.plot(founded_df['founded'], founded_df['growth_percentage'])
plt.xlabel('Founded Year')
plt.ylabel('Average Growth Rate')
plt.title('Average Growth Rate for each Founded Year')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x= 'employees', y= 'estimated_revenues', data=df)
plt.xlabel('Employees')
plt.ylabel('Estimated Revenues')
plt.title('Compares Number of Employees to Estimated Revenues for each Company')
plt.show()
```

```python
sns.histplot(x= 'founded', data=df)
plt.xlabel('Founded')
plt.ylabel('Count')
plt.title('Number of Companies Founded in Each Year')
plt.show()
```
