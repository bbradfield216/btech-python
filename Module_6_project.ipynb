### Which customer has the most invoices

According to the results of the code below, after correcting the standardizatoin errors you can see that 'Stuff by Stew' had the highest number of invoices at 26,509. 'Terry's Trinket had the second most invoices listed at 24,511.  

```python
# Identifies standardization errors
df['Customer'].unique()
```

```python
# Identifies standardization errors
df['Customer'].value_counts()
```

```python
# Corrects the standardization errors from above
# Shows which customer created the most invoices
df['Customer'].replace({"Terry's Trinkets": "Terry's Trinket", "Stuf by Stew": "Stuff by Stew", "stuff by stew": "Stuff by Stew"})
df['Customer'] = df['Customer'].replace({"Terry's Trinkets": "Terry's Trinket", "Stuf by Stew": "Stuff by Stew", "stuff by stew": "Stuff by Stew"})
df['Customer'].value_counts()
```

From this analyis, after correcting the standardization errors, I can conlude that 'Stuff by Stew' had the most invoices.

### Shows some data cleaning steps

```python
#  Based on the number of null values in the column and the questions being asked by executives, this code drops the column 'LastEditedWhen'
df = df.drop(columns='LastEditedWhen')
```

```python
df.info()
```

```python
# Shows how many rows have both a null StockItemID and a null Description
count_both_null = df[(df['StockItemID'].isnull()) & (df['Description'].isnull())].shape[0]
count_both_null
```

```python
# Drops the rows with null values in the 'StockItemID', and 'Description' columns and shows the number of rows left in the data frame
df.dropna(subset=['StockItemID', 'Description'], how='all', inplace=True)
df.shape
```

### Corrects business logic errors in the data

```python
# Turns the negative tax rates into positive tax rates by multiplying by -1
df.loc[df['TaxRate'] < 0, 'TaxRate'] *= -1
```

```python
# Shows the average tax rate
df['TaxRate'].mean()
```

### Imputes missing values in the 'TaxAmount' and 'ExtendedPrice' columns

```python
# Imputes the missing values using the formuala ... tax amount=(unit price * quantity * tax rate) / 100
df['TaxAmount'] = (df['UnitPrice'] * df['Quantity'] * df['TaxRate']) / 100
```

```python
# Shows the average tax amount after removing the null values
df['TaxAmount'].mean()
```

```python
# Imputes the missing values using the formula ... extended price=(unit price * quantity) + tax amount
df['ExtendedPrice'] = (df['UnitPrice'] * df['Quantity']) + df['TaxAmount']
```

```python
# Shows the average extended price after removing the null values from that column
df['ExtendedPrice'].mean()
```

### Shows outliers in the 'Quantity' and 'ExtendedPrice' columns using the z-score method

```python
'''
The below 3 codes shows the process to find outliers in the quanity column using the z-score method
'''
```

```python
average_q = df['Quantity'].mean()
std_dev_q = df['Quantity'].std()
```

```python
df['Z-ScoreQ'] = (df['Quantity'] - average_q) / std_dev_q
```

```python
df.loc[ df['Z-ScoreQ'] >= 3 ]
```

```python
'''
The below 3 codes shows the process to find outliers in the extended price column using the z-score method
'''
```

```python
average_ep = df['ExtendedPrice'].mean()
std_dev_ep = df['ExtendedPrice'].std()
```

```python
df['Z-ScoreEP'] = (df['ExtendedPrice'] - average_ep) / std_dev_ep
```

```python
df.loc[df['Z-ScoreEP'] > 3]
```

### Shows dropping duplicated rows

```python
# Located all duplicated rows sorting by the 'InvoiceID' column
df.loc[ df.duplicated() ].sort_values(by='InvoiceID')
```

```python
# Drops all duplicates and shows the amount of rows left
df.drop_duplicates(inplace=True)
df.shape
```

### Which item has the highest total quantity sold

According to the results of the code below, item 'Black and orange fragile despatch tape 48mmx75m' had the highest total quantity sold.

```python
# Shows which item has the highest total quantity sold
total_quantity_per_item = df.groupby('Description')['Quantity'].sum()
max_quantity_item = total_quantity_per_item.idxmax()
print(max_quantity_item)
```

From this analysis I can conclude that item 'Black and orange fragile despatch tape 48mmx75m' had the highest total quantity sold.

### Which item has generated the most total profit

According to the results of the code below, item '20 mm Double sided bubble wrap 50m' generated the most total profit.

```python
# Shows which item has generated the most total profit
total_lp_per_item = df.groupby('Description')['LineProfit'].sum()
max_lp_item = total_lp_per_item.idxmax()
print(max_lp_item)
```

From this analysis I can conclude that item '20 mm Double sided bubble wrap 50m' generated the most total profit.

### Which item do customers purchase the most quantity of, on average?

According to the results of the code below, item 'Black and orange fragile despatch tape 48mmx75m' was the item customers purchase the most quantity of, on average.


```python
# Shows which item is purchased in greatest quantities in a single order, on average
average_q = df.groupby('Description')['Quantity'].mean()
max_average_item = average_q.idxmax()
print(max_average_item)
```

From this analysis I can conclude that item 'Black and orange fragile despatch tape 48mmx75m' was the item customers purchase the most quantity of, on average.
