```python
import pandas as pd
df = pd.read_csv('invoices.csv')
```

```python
df.head()
```

```python
# Shows which customer created the most invoices
df['Customer'].unique()
```

```python
df = df.drop(columns='LastEditedWhen')
```

```python
df.info()
```

```python
# Shows how many rows have both a null StockItemID and a null Description
count_both_null = df[(df['StockItemID'].isnull()) & (df['Description'].isnull())].shape[0]
count_both_null
```

```python
df.dropna(subset=['StockItemID', 'Description'], how='all', inplace=True)
```

```python
df.loc[df['TaxRate'] < 0, 'TaxRate'] *= -1
```

```python
df['TaxRate'].mean()
```

```python
df['TaxAmount'] = (df['UnitPrice'] * df['Quantity'] * df['TaxRate']) / 100
df['TaxAmount'].mean()
```

```python
df['TaxAmount'] = df['TaxAmount'].fillna((df['UnitPrice'] * df['Quantity'] * df['TaxRate']) / 100)
df['TaxAmount'].mean()
```

```python
df['ExtendedPrice'] = (df['UnitPrice'] * df['Quantity']) + df['TaxAmount']
df['ExtendedPrice'].mean()
```

```python
average_q = df['Quantity'].mean()
std_dev_q = df['Quantity'].std()
```

```python
df['Z-ScoreQ'] = (df['Quantity'] - average_q) / std_dev_q
```

```python
df.loc[ df['Z-ScoreQ'] >= 3 ]
```

```python
average_ep = df['ExtendedPrice'].mean()
std_dev_ep = df['ExtendedPrice'].std()
```

```python
df['Z-ScoreEP'] = (df['ExtendedPrice'] - average_ep) / std_dev_ep
```

```python
df.loc[df['Z-ScoreEP'] > 3]
```

```python
df.loc[ df.duplicated() ].sort_values(by='InvoiceID')
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.groupby('Description')['Quantity'].value_counts(ascending= False)
```

```python
df.groupby('Description')['LineProfit'].value_counts(ascending= False)
```

```python
average_q = df.groupby('Description')['Quantity'].mean()
max_average_item = average_q.idxmax()
print(max_average_item)
```
