```python
# average points of the tennis players below
top_3_greats = [
         { 'name': 'Novak Djokovic', 'points': 11015 },
         {'name': 'Daniil Medvedev', 'points': 10125},
         {'name': 'Alexander Zverev', 'points': 7780}
]


total_points = 0
count = 0

for great in top_3_greats:
    total_points += great['points']
    count += 1

average_points = total_points / count if count > 0 else 0

print(average_points)
```
