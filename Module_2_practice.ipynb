```python
import pandas as pd
df = pd.read_csv('imdb_movies.csv')
```

```python
df.head()
```

```python
# average runtime for movies
df['Runtime'].mean().round(2)
```

```python
# amount of time to watch all movies in hours
hours_time = df['Runtime'].sum() /60
hours_time.round(2)
```

```python
# median Meta_score
df['Meta_score'].median()
```

```python
# most listed genre
df['Genre'].value_counts()
```

```python
# number of different directors
df['Director'].nunique()
```

```python
# actor that appeared the most times
df['Star1'].value_counts()
```

```python
# year oldest movie was released
df['Released_Year'].min()
```

```python
# which movie has the highest gross
df.sort_values(by='Gross', ascending=False)
```

```python
# which movie had the most votes
df.sort_values('No_of_Votes', ascending=False)
```

```python
# which movie had the best rating
df.sort_values('IMDB_Rating', ascending= False)
```

```python
# year with most movie titles
df['Released_Year'].value_counts()
```

```python
# average gross across all movies
avg_gross = df['Gross'].mean()
avg_gross.round(2)
```

```python
# median gross across all movies
df['Gross'].median()
```

```python
# gross profit from the movies in this data set left, right, or no skew
import numpy as np
from scipy.stats import skew

df['Gross'].skew()
```

```python
# creating a new column called 'Gross Per Minute'
df['Gross_Per_Min'] = df['Gross'] / df['Runtime'].round(2)
df['Gross_Per_Min'] = df['Gross_Per_Min'].round(2)
```

```python
# which movie made the most per minute
df.sort_values('Gross_Per_Min', ascending= False)
```
