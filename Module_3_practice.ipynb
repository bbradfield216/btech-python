```python
import pandas as pd
df = pd.read_csv('horse_racing.csv')
```

```python
df.head()
```

```python
# average weight of all jockey's in kg
round(df['Jockey weight'].mean(), 2)
```

```python
# average race time for brett prebble
brett_prebble = df[df['Jockey'] == 'Brett Prebble']
round(brett_prebble['Race time'].mean(), 2)
```

```python
# how much money first place jockeys won, on average
first_place = df[df['Final place'] == 1]
round(first_place['Prize money'].mean(), 2)
```

```python
# surface winning horses tend to run on
first_place['Surface'].value_counts()
```

```python
# fastest race time among horses that ran the 1400 meter race
distance = df[df['Distance'] == 1400]
distance['Race time'].min()
```

```python
# slowest race time among Irish horses who ran the 1400 meter race
distance = df[(df['Distance'] == 1400) & (df['Country'] == 'Irland')]
distance['Race time'].max()
```

```python
# australian jockey competed in  most races
australian_jockey = df[df['Country'] == 'Australia']
australian_jockey['Jockey'].value_counts()
```

```python
# jockey with the least likelihood (highest number of odds) from australia, ireland, or new zealand
three_certain_countries = df[(df['Country'] == 'Australia') | (df['Country'] == 'Irland') | (df['Country'] == 'New Zealand')]
three_certain_countries[['Odds', 'Jockey']].groupby('Jockey').mean().sort_values(by='Odds', ascending= False)
```

```python
# horse id age 10 or above with the best average race time (smallest)
horse_age = df[df['Horse age'] >= 10]
horse_age[['Race time', 'HorseId']].groupby('HorseId').mean().sort_values(by= 'Race time')
```

```python
# country who  won the most races (first place) where the prize money was over $2,000,000
first_and_over_one_mill = df[(df['Final place'] == 1) & (df['Prize money'] > 2000000)]
first_and_over_one_mill['Country'].value_counts()
```
